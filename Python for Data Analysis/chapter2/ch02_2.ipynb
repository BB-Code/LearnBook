## MovieLeans 1M 数据集

```python
import pandas as pd
import pylab
```

```python
unames =['user_id','gender','age','occupation','zip']
print(unames)
```

```python
users = pd.read_table('movielens/users.dat',sep='::',header=None,names=unames)
```

```python
ranmes =['user_id','movie_id','rating','timestamp']
```

```python
ratings = pd.read_table('movielens/ratings.dat',sep='::',header=None,names=ranmes)
```

```python
mnames =['movie_id','title','genres']
```

```python
movies = pd.read_table('movielens/movies.dat',sep='::',header=None,names=mnames)
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

## 合并数据
- 与表之间要有相关联字段

```python
data = pd.merge(pd.merge(ratings,users),movies)
```

```python
type(data)
```

```python
data.ix[0]
```

```python
mean_ratings = data.pivot_table('rating','title',columns='gender',aggfunc='mean')
print(mean_ratings)
```

```python
print(help(data.pivot_table))
```

- 去掉评分数据不够250条的电影

```python
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title[:5]
```

```python
actives_titles = ratings_by_title.index[ratings_by_title>=250]
```

```python
actives_titles[:5]
```

```python
mean_ratings = mean_ratings.ix[actives_titles]
```

```python
mean_ratings
```

- 女性最喜欢的电影

```python
top_female_ratings = mean_ratings.sort_index(by='F',ascending=False)
```

```python
top_female_ratings[:10]
```

- 计算评分分歧

```python
mean_ratings['diff'] = mean_ratings['M']-mean_ratings['F']
```

```python
mean_ratings
```

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
```

```python
sorted_by_diff[:10]
```

```python
sorted_by_diff[::-1][:10] #男性更喜欢的电影
```

- 不考虑性别找出分歧最大的电影,可计算数据的方差和标准差

```python
#标准差
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
#过滤掉评论少于250的数据
rating_std_by_title = rating_std_by_title.ix[actives_titles]
```

```python
#降序排列
rating_std_by_title.order(ascending=False)[:10]
```
