## 1880-2010 年间全美婴儿姓名

```python
import pandas as pd
import numpy as np
import pylab
```

```python
name1880 = pd.read_csv('babynames/yob1880.txt',names=['name','sex','births'])
```

```python
name1880
```

```python
name1880.groupby('sex').births.sum()
```

```python
years = range(1880,2011)
```

```python
columns = ['name','sex','births']
pieces =[]
```

```python
for year in years:
    path = 'babynames/yob%d.txt' %year
    frame = pd.read_csv(path,names=columns)
    frame['year'] = year
    pieces.append(frame)
```

```python
names = pd.concat(pieces,ignore_index=True)
```

```python
names.tail(10)
```

```python
names.count
```

```python
total_births = names.pivot_table('births','year',columns='sex',aggfunc=sum)
```

```python
total_births.tail(5)
```

```python
%pylab
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births/births.sum()
    return group
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
```

```python
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
```

```python
def get_top_1000(group):
    return group.sort_index(by='births',ascending=False)[:1000]
```

```python
grouped = names.groupby(['year','sex'])
```

```python
top1000 = grouped.apply(get_top_1000)
```

```python
pieces =[]
```

```python
for year,group in names.groupby(['year','sex']):
    pieces.append(group.sort_index(by='births',ascending=False)[:1000])
```

```python
top1000  = pd.concat(pieces,ignore_index=True)
```

```python
top1000
```

```python
def get_quantile_count(group,q=0.5):
    group = group.sort_index(by='prop',ascending=False)
    return group.prop.cumsum().searchsorted(q)[0]+1
```

```python
diversity = top1000.groupby(['year','sex']).apply(get_quantile_count)
```

```python
diversity = diversity.unstack('sex')
```

```python
diversity.head(10)
```

```python
diversity.plot(title='NUmber of popular names in top 50%')
```

## 最后一个字母的变革

```python
get_last_letter = lambda x:x[-1]
```

```python
last_letters = names.name.map(get_last_letter)
last_letters
```

```python
names['last_letter'] = last_letters
names
```

```python
table  = names.pivot_table('births','last_letter',columns=['sex','year'],aggfunc=sum)
```

```python
subtable = table.reindex(columns=[1910,1960,2010],level='year')
```

```python
subtable.head(5)
```

```python
subtable.sum()
```

```python
letter_prop = subtable/subtable.sum().astype(float)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes = plt.subplots(2,1,figsize=(10,8))
letter_prop['M'].plot(kind='bar',rot=0,ax=axes[0],title='Male')
letter_prop['F'].plot(kind='bar',rot=0,ax=axes[1],title='Female',legend=False)
```

- 变成女孩名字的男孩名字（以及相反情况）

```python
all_names = top1000.name.unique()
```

```python
all_names
```

```python
mask = np.array(['lesl' in x.lower() for x in all_names])
```

```python
lesley_like =  all_names[mask]
```

```python
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
```

```python
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births','year',columns='sex',aggfunc='sum')
```

```python
table.head()
```

```python
table = table.div(table.sum(1),axis=0)
```

```python
table.tail()
```
